### 1. How many companies/founders are funded in total?

```python
total_companies = len(df)
total_founders = df["total_founders"].sum()
print("Total companies: {}. Total founders: {}.".format(total_companies, total_founders))
```

### 2. What is the distribution of companies by sector?

```python
# replace companies
print("Total sectors: {}".format(len(df["company_cat"].unique())))
df["company_cat"].fillna("Not given", inplace=True)
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(16, 6))
sns.countplot(x="company_cat", data=df, order=df["company_cat"].value_counts().index)
plt.xticks(rotation=45, ha="right")
plt.xlabel("Sector")
plt.ylabel("Count")
plt.title("Distribution of companies by sector")
plt.show()
```

#### Top 3 funded sectors are:
<ol>
    <li>Finance</li>
    <li>Enterprise Solutions</li>
    <li>Digital Health</li>
</ol>

### 3. What is the distribution of companies by cohort/location?

```python
sns.set(style="darkgrid")
plt.figure(figsize=(16, 6))
sns.countplot(x="company_cohort", data=df, order=df["company_cohort"].value_counts().index)
plt.xticks(rotation=45, ha="right")
plt.xlabel("Cohort")
plt.ylabel("Count")
plt.title("Distribution of companies by cohort")
plt.show()
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(16, 6))
sns.countplot(x="company_location", data=df, order=df["company_location"].value_counts().index)
plt.xticks(rotation=45, ha="right")
plt.xlabel("Location")
plt.ylabel("Count")
plt.title("Distribution of companies by location")
plt.show()
```
